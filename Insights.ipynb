```python
import pandas as pd
import numpy as np

# Google Drive shared file link
url1 = 'https://drive.google.com/file/d/15OY8NtKdOV4T7fHWMie_tAkYt3NnEjM7/view?usp=drive_link'

# Extract the file ID from the URL
file_id1 = url1.split('/')[-2]

# Construct the direct download URL
dwn_url1 = f'https://drive.google.com/uc?id={file_id1}'

# Load the CSV with a compatible encoding
df1 = pd.read_csv(dwn_url1, encoding='ISO-8859-1')

# Show the first few rows
df1.head()
```

```python
import pandas as pd

# Assuming df1 is already loaded and 'TI' and 'S' are numeric
# Sort by 'TI' and 'S' both in descending order
top10 = df1.sort_values(by=['TI', 'S'], ascending=[False, False]).head(10)

# Display the top 10 rows
print(top10[['Content','Url','Posted At','Comments_Sentiment','Sentiment React','TI', 'S']])
```

```python
# Filter and print only the viral posts
viral_only = df1[df1['Viral'] == True]

# Display selected columns, or all if you prefer
viral_only[['Content', 'TI_z', 'S_z', 'Url', 'Posted At']]
```

```python
import matplotlib.pyplot as plt

# Count the number of True and False values
viral_counts = df1['Viral'].value_counts()

# Labels and colors for better visuals
labels = ['Not Viral (False)', 'Viral (True)']
colors = ['#ff9999', '#66b3ff']

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(viral_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Viral vs Non-Viral Posts')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle
plt.show()
```

```python
import pandas as pd

# Clean and parse datetime
df1['Posted At'] = df1['Posted At'].astype(str).str.strip()
df1['Posted At'] = pd.to_datetime(df1['Posted At'], errors='coerce')
df1 = df1.dropna(subset=['Posted At'])

# Extract date
df1['Post Date'] = df1['Posted At'].dt.date

# Count posts per day
daily_counts = df1.groupby('Post Date').size().reset_index(name='Posts Per Day')

# Basic statistics
count_days = daily_counts['Posts Per Day'].count()
mean_posts = daily_counts['Posts Per Day'].mean()
median_posts = daily_counts['Posts Per Day'].median()
mode_posts = daily_counts['Posts Per Day'].mode()
min_post_day = daily_counts.loc[daily_counts['Posts Per Day'].idxmin()]
max_post_day = daily_counts.loc[daily_counts['Posts Per Day'].idxmax()]
std_dev = daily_counts['Posts Per Day'].std()
variance = daily_counts['Posts Per Day'].var()
range_posts = max_post_day['Posts Per Day'] - min_post_day['Posts Per Day']

# IQR and outlier bounds
Q1 = daily_counts['Posts Per Day'].quantile(0.25)
Q3 = daily_counts['Posts Per Day'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Outliers
outliers = daily_counts[(daily_counts['Posts Per Day'] < lower_bound) |
                        (daily_counts['Posts Per Day'] > upper_bound)]

# Display results
print(" Full Statistics of Posts Per Day:")
print(f"Total Days Recorded: {count_days}")
print(f"Mean: {mean_posts:.2f}")
print(f"Median: {median_posts}")
print(f"Mode: {', '.join(map(str, mode_posts.values))}")
print(f"Min: {min_post_day['Posts Per Day']} on {min_post_day['Post Date']}")
print(f"Max: {max_post_day['Posts Per Day']} on {max_post_day['Post Date']}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Range (Max - Min): {range_posts}")
print(f"IQR (Q3 - Q1): {IQR}")
print(f"Outlier Lower Bound: {lower_bound}")
print(f"Outlier Upper Bound: {upper_bound}")
print(f"Number of Outlier Days: {len(outliers)}")

# Optional: show outlier dates
if not outliers.empty:
    print("\n Outlier Days:")
    print(outliers.sort_values(by='Posts Per Day', ascending=False).to_string(index=False))
else:
    print("\nNo outliers detected.")
```
